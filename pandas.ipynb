```python
import pandas as pd
data={
      '영화':['명량','극한직업','신과함께','국제시장','괴물','도둑들','7번방의 선물','암살'],
      '개봉연도':[2014,2019,2017,2014,2006,2012,2013,2015],
      '관객 수':[1761,1626,1441,1426,1301,1298,1281,1270],
      '평점':[8.88,9.20,8.73,9.16,8.62,7.64,8.83,9.10]
}
df=pd.DataFrame(data)
df
```

```python
df['영화']
```

```python
df[['영화','평점']]
```

```python
#2015년 이후에 개봉한 영화중 영화 개봉연도 정보만 출력
recent=df['개봉연도']>=2015
df.loc[recent,['개봉연도','영화']]
```

```python
#추천점수 column추가
#추천점수=(관객수*평점)//100
df['추천점수']=(df['관객 수']*df['평점'])//100
df
```

```python
#전체 데이터를 개봉연도 기준 내림차순으로 출력
df['개봉연도'].sort_values(ascending=False)
```
